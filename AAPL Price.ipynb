```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
df = pd.read_csv('aapl_price_v5.csv')
print(df)
```

```python
df.columns = ['Date', 'Open','Close','Low','High','Volume']
# plt.scatter(
#     x = df['Date'], y = df['Price']
# )
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['Open'], color='blue')
plt.ylabel('$')
plt.title('AAPL Open Price')

plt.subplot(1, 2, 2)
plt.plot(df['Close'], color='green')
plt.ylabel('$')
plt.title('AAPL Close Price')
plt.tight_layout()
plt.show()
```
