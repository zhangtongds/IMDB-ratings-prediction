```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
df = pd.read_csv('aapl_price_v5.csv')
print(df)
```

```python
df.columns = ['Date', 'Open','Close','Low','High','Volume']
# plt.scatter(
#     x = df['Date'], y = df['Price']
# )
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['Open'], color='blue')
plt.xlabel('Sequence')
plt.ylabel('Open Price')
plt.title('AAPL Open Price')

plt.subplot(1, 2, 2)
plt.plot(df['Close'], color='green')
plt.ylabel('$')
plt.title('AAPL Close Price')
plt.tight_layout()
plt.show()
```

```python
plt.scatter(np.linspace(0,df['Open'].size, df['Open'].size), y = df['Open'])
plt.xlabel('Sequence')
plt.ylabel('Open Price')
plt.title('AAPL Open Price')
plt.show()
```

```python
plt.hist(df['Open'], color='red', bins=np.arange(25, 120), alpha=.5)
plt.xlabel('Sequence')
plt.ylabel('Open Price')
plt.title('AAPL Open Price')
```

```python
plt.boxplot(df['Open'])
plt.xlabel('Sequence')
plt.ylabel('Open Price')
plt.title('AAPL Open Price')
```
